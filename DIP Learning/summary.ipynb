2.多图片布局

````python
img = cv2.imread("data/fruits.bmp")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(2,2,1)
plt.imshow(img)

# according to the designated width and height
res = cv2.resize(img, (300, 100))
# according to the scale, for example: 2 times in x and y direction
res2 = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)

plt.subplot(2,2,2)
plt.imshow(res)
plt.subplot(2,2,3)
plt.imshow(res2)
````

##### （二）图像的几何变化

1.resize()  改变大小和channel
* ```cv2.resize(filename, methods) ```

five interpolation methods：

> * ```cv2.INTER_NEAREST ```Nearest neighbor interpolation最近邻插值
> * ```cv2.INTER_LINEAR ```linear interpolation线性插值
> * ```cv2.INTER_AREA``` local pixel resampling, regional interpolation区域插值
> * ```cv2.INTER_CUBIC``` Cubic interpolation based on 4x4 neighborhood基于4x4邻域的三次插值
> * ```cv2.INTER_LANCZOS4 ```Lanczos Interpolation based on 8x8 neighborhood基于8x8邻域的Lanczos插值


````python
# according to the designated width and height
res = cv2.resize(img, (300, 100))
# according to the scale, for example: 2 times in x and y direction根据刻度，例如:x和y方向上的2倍
res2 = cv2.resize(img, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
````
